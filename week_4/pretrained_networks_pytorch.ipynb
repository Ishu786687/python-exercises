```python
from torchvision import models
from torchvision import transforms
from PIL import Image
import torch
import pandas as pd
```

```python
dir(models)
```

```python
resnet = models.resnet50(pretrained=True)
```

```python
resnet
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])
    ])
```

```python
img = Image.open("C:\\pytorch\\assets\\dog1.jpg")
img.show()
```

```python
image_tensor = preprocess(img)
```

```python
batch_img = torch.unsqueeze(image_tensor, 0)
```

```python
resnet.eval()
```

```python
output = resnet(batch_img)
```

```python
json_path = 'C:\\pytorch\\assets\\imagenet_class_index.json'
# Read the JSON data into a DataFrame
labels = pd.read_json(json_path)
# Display the first few rows of the DataFrame
labels.head()
```

```python
_, index = torch.max(output,1)
```

```python
percentage = torch.nn.functional.softmax(output, dim=1)[0] * 100
labels[index], percentage[index[0]].item()
```
